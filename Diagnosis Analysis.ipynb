```python
#Importing all important libraries

import pandas as pd
import numpy as np
```

```python
#Loading Training Data.

train = pd.read_csv("Data/training.csv")
pd.set_option('display.max_columns', None)
train.head()
```

```python
#Loading Testing Data.

test = pd.read_csv("Data/testing.csv")
pd.set_option('display.max_columns', None)
test.head()
```

```python
train = train.drop(["Unnamed: 133"], axis = 1)
```

```python
train_duplicates = train[train.duplicated()]
train_duplicates.shape
```

```python
test_duplicates = test[test.duplicated()]
test_duplicates.shape
```

```python
#labeling the categorical data

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
train["prognosis"] = le.fit_transform(train["prognosis"])
test["prognosis"] = le.fit_transform(test["prognosis"])
```

```python
#Scaling the data between 0 and 1

train = (train-train.min())/(train.max()-train.min())
test = (test-test.min())/(test.max()-test.min())
```

```python
train.tail()
```

```python
train["prognosis"].unique()
```

```python
test["prognosis"].unique()
```

```python
test.head()
```

```python
#Spliting data into train and test

x_train = train.drop(columns = "prognosis")
y_train = train["prognosis"]
x_test = test.drop(columns = "prognosis")
y_test = test["prognosis"]
```

```python
x_train = x_train.dropna(axis = 1)
x_test = x_test.dropna(axis = 1)
print("x_train", x_train.shape)
print("x_test", x_test.shape)
```

```python
y_train.shape
```

```python
x_test.head()
```

```python
#KNeighborsRegressor model

from sklearn.neighbors import KNeighborsRegressor
ne = KNeighborsRegressor(n_neighbors=3)
ne.fit(x_train, y_train)
```

```python
#Accuracy of the model.
Score = ne.score(x_test, y_test)*100
```

```python
print("KNeighborsRegressor Score:", Score,"%")
```

```python
#DecisionTreeRegressor Model

from sklearn.tree import DecisionTreeRegressor
dr = DecisionTreeRegressor(criterion = 'squared_error')
dr.fit(x_train, y_train)
print("DecisionTreeRegressor Score:", dr.score(x_test, y_test)*100)
```

```python
#LinearRegression Model.

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
print("LinearRegression Score:", lr.score(x_test, y_test)*100)
```
